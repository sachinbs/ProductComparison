```python
##Function to collect twitter data

def collecttwittertweets(consumer_key, consumer_secret, searchquery):
    consumer_key = consumer_key
    consumer_secret = consumer_secret
    # Replace the API_KEY and API_SECRET with your application's key and secret.
    auth = tweepy.AppAuthHandler(consumer_key, consumer_secret)

    api = tweepy.API(auth, wait_on_rate_limit=True,
                       wait_on_rate_limit_notify=True)

    if (not api):
        print ("Can't Authenticate")
        sys.exit(-1)
    searchQuery = searchquery  # this is what we're searching for
    maxTweets = 10000000 # Some arbitrary large number
    tweetsPerQry = 100  # this is the max the API permits
    fName = 'collected_twitter_tweets.txt' # We'll store the tweets in a text file.

    # If results from a specific ID onwards are read, set since_id to that ID.
    # else default to no lower limit, go as far back as API allows
    sinceId = None

    # If results only below a specific ID are, set max_id to that ID.
    # else default to no upper limit, start from the most recent tweet matching the search query.
    max_id = 0

    tweetCount = 0
    print("Downloading max {0} tweets".format(maxTweets))
    with open(fName, 'a') as f:
        while tweetCount < maxTweets:
            try:
                ## using the tweet_mode to be extended in order to collect tweets with full text
                if (max_id <= 0):
                    if (not sinceId):
                        new_tweets = api.search(q=searchQuery, count=tweetsPerQry,tweet_mode= 'extended')
                    else:
                        new_tweets = api.search(q=searchQuery, count=tweetsPerQry,tweet_mode = 'extended',
                                                since_id=sinceId)
                else:
                    if (not sinceId):
                        new_tweets = api.search(q=searchQuery, count=tweetsPerQry,tweet_mode= 'extended',
                                                max_id=str(max_id - 1))
                    else:
                        new_tweets = api.search(q=searchQuery, count=tweetsPerQry,tweet_mode= 'extended',
                                                max_id=str(max_id - 1),
                                                since_id=sinceId)
                if not new_tweets:
                    print("No more tweets found")
                    break
                for tweet in new_tweets:  ##collecting tweets which has the placw value in it
                    if tweet.place!= None:
                        f.write(jsonpickle.encode(tweet._json, unpicklable=False) +'\n')
                tweetCount += len(new_tweets)
                print("Downloaded {0} tweets".format(tweetCount))
                max_id = new_tweets[-1].id
            except tweepy.TweepError as e:
                # Just exit if any error
                print("some error : " + str(e))
                break

    print ("Downloaded {0} tweets, Saved to {1}".format(tweetCount, fName))
```

```python
## Load the text file and convert all text into lower case for easy processing of content

def dataprocessing(textfilename):
    tweets_data = []
    tweets_file = open(textfilename, "r")
    for line in tweets_file:
        try:
            tweet = json.loads(line.lower())
            tweets_data.append(tweet)
        except:
            continue
    pd.set_option('display.max_colwidth', -1)  ## To set maximum column size to the data frame c
    common = pd.DataFrame()  ## Creating a new data frame which contains the entire test data
    tweetslist = []
    tweetscountry = []
     
    ## Analyse the test data and create two new columns for the tweets and its respective location
    for i in range(len(tweets_data)):
        tweetslist.append(tweets_data[i]['full_text'])
    # If place value is not none, then we can assign the country to the data frame
    # If not we go ahead and assign the user's defined location to the column
    # Most times location value in our tweets does not yeild us useful data
        if tweets_data[i]['place']!= None:
            tweetscountry.append(tweets_data[i]['place']['country'])
        else:
            tweetscountry.append(tweets_data[i]['user']['location'])
    common['Tweet_Text'] = tweetslist
    common['Location'] = tweetscountry
    return common
```

```python
# Sentiment analysis using textblob
def clean_tweet(tweet):
    '''
    Utility function to clean the text in a tweet by removing 
    links and special characters using regex.
    '''
    tweet1 = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split())
    return tweet1.lower()

def analize_sentiment(tweet):
    '''
    Utility function to classify the polarity of a tweet
    using textblob.
    '''
    analysis = TextBlob(clean_tweet(str(tweet)))
    if analysis.sentiment.polarity > 0:
        return 1
    elif analysis.sentiment.polarity == 0:
        return 0
    else:
        return -1
```

```python
## Performing NLTK to tokenize every tweet and extract only the nouns and verbs from the attributes dictionary
def performNLTK(placeholder1,placeholder2):
    from nltk.corpus import stopwords
    from nltk.tokenize import word_tokenize
    
    ## All possible attributes related to any electronic products

    attributes_dict = {
        'performance': ['performance', 'performs', 'performing', 'speed'],
        'service': ['service', 'services', 'servicing'],
        'design': ['design', 'designs', 'body'],
        'battery': ['battery', 'batteries', 'backup', 'juice'],
        'camera': ['camera', 'cameras'],
        'os': ['os'],
        'memory': ['memory', 'storage', 'space'],
        'touch': ['touch'],
        'display': ['display', 'screen'],
        'price': ['price', 'cost', 'money'],
        'innovation': ['innovation', 'innovative', 'unique'],
        'resistance': ['resistance']
    }
    words_list = ['performance','service','design','battery','camera','OS','memory','touch','display','price',
                  'innovation', 'update', 'resistance']

    placeholders = [placeholder1.lower(), placeholder2.lower()]
    
    ## Adding two new columns to place the tokenized words and the product company
    common['words'] = ""
    common['product'] = None
    filtered_sentence = []
    for i in range(len(common)):
        word_tokens = word_tokenize(common['Tweet_Text'][i])
        for w in word_tokens:
            if w in placeholders:
                common['product'][i] = w.lower()
            for attr in attributes_dict:
                if w in attributes_dict[attr]:
                    filtered_sentence.append(attr)
        if not filtered_sentence:
            common['words'][i] = None
        else:
            common['words'][i] = filtered_sentence

        filtered_sentence = []
    return common
```

```python
## Creating unique positive words
def positivewords(data):
    pos_dict = []
    for i in range(len(data)):
        if data['SA'][i] == 1:
            for x in data['words'][i]:
                pos_dict.append(x)
    # Creating a dictionary of frequencies of words
    pos_frequencies = []
    for w in pos_dict:
        pos_frequencies.append(pos_dict.count(w))
    pos_freq = {}
    for i, j in zip(pos_dict, pos_frequencies):
        pos_freq[i] = j

    # Removes elements from the list "pos_dict" if frequency is 3 or less than 3
    pos_words = pos_dict
    for i in pos_freq:
        if pos_freq[i] < 4:
            pos_words = [y for y in pos_words if y != i]
    pos_dict_unique = set(pos_words)
    pos_dict_unique = list(pos_dict_unique)
    return pos_dict_unique
```

```python
## Sankey diagram label list for processing

def sankeydiagramlabellist(placeholder):
    labellist = []
    labellist.append(placeholder)
    labellist.append("positive")
    labellist.append("negative")
    labellist.append("neutral")
    
    
    ## Calls the positivewords function which will give the unique list of positive words
    
    placeholder_pos_dict_unique = positivewords(placeholderdf(placeholder))
    length_pos = len(placeholder_pos_dict_unique)
    
    ## Appends all positive words to the list
    for i in placeholder_pos_dict_unique:
        labellist.append(i)
    
    ## Calls the negativewords function which will give the unique list of negative words
    
    placeholder_neg_dict_unique = negativewords(placeholderdf(placeholder))
    length_neg = len(placeholder_neg_dict_unique)
    
    ## Appends all negative words to the list
    for i in range(length_neg):
        labellist.append(placeholder_neg_dict_unique[i])
        
     ## Calls the neutralwords function which will give the unique list of neutral words
        
    placeholder_neu_dict_unique = neutralwords(placeholderdf(placeholder))
    length_neu = len(placeholder_neu_dict_unique) 
    
     ## Appends all neutral words to the list
    for i in range(length_neu):
        labellist.append(placeholder_neu_dict_unique[i])

    labellist = sorted(set(labellist), key=labellist.index)
    return labellist 
```

```python

## A sankey diagram is a visualization used to depict a flow from one set of values to another 
## The things being connected are called nodes and the connections are called links
## Sankeys are best used when you want to show a many-to-many mapping between two domains

def sankeydiagram(placeholder):
    sourcelist = [0,0,0]
    placeholder_pos_dict_unique = positivewords(placeholderdf(placeholder))
    placeholder_neg_dict_unique = negativewords(placeholderdf(placeholder))
    placeholder_neu_dict_unique = neutralwords(placeholderdf(placeholder))

    for i in range(len(placeholder_pos_dict_unique)):
        sourcelist.append(1)
    for i in range(len(placeholder_neg_dict_unique)):
        sourcelist.append(2)
    for i in range(len(placeholder_neu_dict_unique)):
        sourcelist.append(3)

    targetlist = [1,2,3]
    valuelist = [8,8,8]
    colorlist = ['rgb(4, 192, 201)',"rgb(155, 232, 29)","Red","Grey"]
    colorlinks = ["rgb(178,223,138)","rgb(248, 27, 62)","rgb(220, 220, 220)"]
    labellist = sankeydiagramlabellist(placeholder)
    for i in placeholder_pos_dict_unique:
        index = labellist.index(i)
        targetlist.append(index)
        valuelist.append(1)
        colorlist.append("rgb(4, 192, 201)")
        colorlinks.append("rgb(178,223,138)")
    for i in placeholder_neg_dict_unique:
        index = labellist.index(i)
        targetlist.append(index)
        valuelist.append(1)
        colorlist.append("rgb(4, 192, 201)")
        colorlinks.append("rgb(248, 27, 62)")
    for i in placeholder_neu_dict_unique:
        index = labellist.index(i)
        targetlist.append(index)
        valuelist.append(1)
        colorlist.append("rgb(4, 192, 201)")
        colorlinks.append("rgb(220, 220, 220)")

    import plotly
    from plotly.graph_objs import Scatter, Layout

    plotly.offline.init_notebook_mode(connected=True)

    data = dict(
        type='sankey',
        node = dict(
          pad = 15,
          thickness = 20,
          line = dict(
            color = "black",
            width = 0.5
          ),
          label = labellist,
          color = colorlist
        ),
        link = dict(
          source = sourcelist,
          target = targetlist,
          value = valuelist,
          color = colorlinks

      ))

    layout =  dict(
        title = "Sankey Diagram:  " + placeholder.title() ,
        font = dict(
          size = 10
        )
    )

    fig = dict(data=[data], layout=layout)
    plotly.offline.iplot(fig)
```
